üß™ import_toy_regression_metrics
Understanding regression metrics & custom scorers

üì¶ 1. Imports de m√©tricas

```python
from sklearn.metrics import mean_squared_error          # Importa MSE (penaliza erros grandes)
from sklearn.metrics import mean_absolute_error         # Importa MAE (erro m√©dio absoluto em ‚Ç¨)
from sklearn.metrics import r2_score                    # Importa R¬≤ (vari√¢ncia explicada)
from sklearn.metrics import make_scorer                 # Importa utilit√°rio para criar scorers customizados

import numpy as np                                      # Importa NumPy para trabalhar com arrays
```

üöó 2. Valores reais e previs√µes (toy Cars4You)

```python
y_true = np.array([7000, 10000, 14500, 21000, 12000])   # Pre√ßos reais dos carros (ground truth)

y_pred_model_a = np.array([7200, 9800, 15000, 20500, 11800])  # Previs√µes do Modelo A (erros pequenos e consistentes)
y_pred_model_b = np.array([7000, 10000, 14500, 26000, 12000]) # Previs√µes do Modelo B (um erro grande)
```

üìè 3. C√°lculo das m√©tricas (MAE, MSE, R¬≤)

```python
mae_a = mean_absolute_error(y_true, y_pred_model_a)     # Calcula MAE do Modelo A
mse_a = mean_squared_error(y_true, y_pred_model_a)      # Calcula MSE do Modelo A
r2_a = r2_score(y_true, y_pred_model_a)                 # Calcula R¬≤ do Modelo A

mae_b = mean_absolute_error(y_true, y_pred_model_b)     # Calcula MAE do Modelo B
mse_b = mean_squared_error(y_true, y_pred_model_b)      # Calcula MSE do Modelo B
r2_b = r2_score(y_true, y_pred_model_b)                 # Calcula R¬≤ do Modelo B
```

üìä 4. Compara√ß√£o dos modelos

```python
print("Model A ‚Üí MAE:", mae_a, "| MSE:", mse_a, "| R2:", r2_a)  # Mostra m√©tricas do Modelo A
print("Model B ‚Üí MAE:", mae_b, "| MSE:", mse_b, "| R2:", r2_b)  # Mostra m√©tricas do Modelo B
```

‚öôÔ∏è 5. Cria√ß√£o de scorer customizado (make_scorer)

```python
mae_scorer = make_scorer(                               # Cria um scorer customizado para MAE
    mean_absolute_error,                                # Usa a fun√ß√£o MAE como m√©trica base
    greater_is_better=False                             # Indica que menor MAE = melhor modelo
)                                                       # Fim da defini√ß√£o do scorer
```

üß† Regra pr√°tica Cars4You (para fechar o toy)
MAE ‚Üí erro m√©dio em euros (interpret√°vel)
MSE ‚Üí penaliza erros grandes (outliers)
R¬≤  ‚Üí explica vari√¢ncia, n√£o erro real
make_scorer ‚Üí torna m√©tricas de erro compat√≠veis com CV

üß™ import_toy_debug_metrics_experiment

```python
import numpy as np                                      # Arrays num√©ricos
import pandas as pd                                     # Tabelas para resumo

from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)
```

```python
y_true = np.array([7000, 10000, 14500, 21000, 12000])   # Pre√ßos reais

predictions = {
    "Model_A_stable": np.array([7200, 9800, 15000, 20500, 11800]),  # erros pequenos
    "Model_B_outlier": np.array([7000, 10000, 14500, 26000, 12000]) # 1 erro grande
}
```

```python
metrics = {
    "MAE": mean_absolute_error,          # Erro m√©dio absoluto (‚Ç¨/interpret√°vel)
    "MSE": mean_squared_error,           # Penaliza erros grandes
    "R2": r2_score                       # Vari√¢ncia explicada
}
```

```python
debug_results = []                       # Guardar resultados para resumo final

for model_name, y_pred in predictions.items():

    print("\n" + "=" * 60)
    print(f"üîç Evaluating {model_name}")
    print("=" * 60)

    for metric_name, metric_fn in metrics.items():

        value = metric_fn(y_true, y_pred)            # Calcula a m√©trica
        print(f"{metric_name}: {value:.4f}")         # Print estilo debugger

        debug_results.append({
            "model": model_name,
            "metric": metric_name,
            "value": value
        })
```

```python
debug_df = pd.DataFrame(debug_results)                # Converte resultados em DataFrame
debug_df
```

```python
pivot_df = debug_df.pivot(
    index="model",
    columns="metric",
    values="value"
)

pivot_df
```

üß† 6. O que este experimento mostra (li√ß√£o chave)

- MAE prefere modelos com erros consistentes
- MSE pune fortemente outliers
- R¬≤ pode parecer ‚Äúbom‚Äù mesmo com erros grandes
- A m√©trica escolhida muda completamente a decis√£o
