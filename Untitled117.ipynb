```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno

sns.set(style="whitegrid")
```

```python
df = pd.DataFrame({
    "year": [2015, 2016, 2017, 2018, 2019, 2020],
    "kilometers": [90000, 75000, 60000, 40000, 30000, np.nan],
    "power": [90, 110, 130, 150, 180, 200],
    "fuel": ["diesel", "diesel", "petrol", "petrol", "hybrid", "hybrid"],
    "price": [12000, 13500, 15000, 18500, 22000, 25000]
})

df
```

```python
plt.figure()
plt.hist(df["price"], bins=5)
plt.xlabel("Price (â‚¬)")
plt.ylabel("Count")
plt.title("Distribution of car prices")
plt.show()
```

```python
plt.figure()
sns.scatterplot(
    data=df,
    x="kilometers",
    y="price",
    hue="fuel",
    s=100
)
plt.title("Price vs Kilometers by Fuel Type")
plt.show()
```

```python
# Load Cars4You training data
df = pd.read_csv("train.csv")

print("Dataset shape:", df.shape)
df.head()
```

```python
df.info()
```

```python
plt.figure()
sns.histplot(df["price"], bins=50, kde=True)
plt.xlabel("Price")
plt.title("Distribution of car prices (target)")
plt.show()
```

```python
plt.figure()
sns.scatterplot(
    data=df,
    x="mileage",
    y="price",
    alpha=0.4
)
plt.title("Price vs mileage")
plt.show()
```

```python
plt.figure()
sns.scatterplot(
    data=df,
    x="year",
    y="price",
    alpha=0.4
)
plt.title("Price vs Year")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(
    data=df,
    x="fuelType",
    y="price"
)
plt.xticks(rotation=45)
plt.title("Price distribution by fuel type")
plt.show()
```

```python
msno.matrix(df)
plt.show()
```

```python
msno.bar(df)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
corr = df.select_dtypes(include=np.number).corr()

sns.heatmap(
    corr,
    cmap="coolwarm",
    center=0
)
plt.title("Correlation matrix (numerical features)")
plt.show()
```
